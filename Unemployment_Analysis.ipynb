```python
import pandas as pd
import numpy as np

# graphical libaries
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.figure_factory as ff
import plotly.graph_objects as go
from plotly.subplots import make_subplots

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/content/unemployment analysis.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
country_name = []
country_code = []
year = []
unemployment_rate = []
column_names = df.columns[2:]

for index, row in df.iterrows():
    for temp_year in column_names:
        country_name.append(row['Country Name'])
        country_code.append(row['Country Code'])
        year.append(temp_year)
        unemployment_rate.append(row[temp_year])

df2 = pd.DataFrame(data={'Year': year,'Country Name': country_name,'Country Code': country_code,'Unemployement Rate': unemployment_rate})
df2.head()
```

```python
fig = px.choropleth(df2, locations="Country Code",
                    color="Unemployement Rate",
                    hover_name="Country Name",
                    animation_frame="Year",
                    range_color=(0, 40),
                    color_continuous_scale=px.colors.sequential.Plasma,
                    title=' Map + Time Line')
#fig.update_layout(transition = {'duration': 10})
fig.show()
```

```python
fig = px.bar(df2.groupby('Country Name').mean('Unemployement Rate').sort_values('Unemployement Rate')[:10],x=df2.groupby('Country Name').mean('Unemployement Rate').sort_values('Unemployement Rate').index[:10],
             y='Unemployement Rate',
             color_discrete_sequence=px.colors.qualitative.Set3,
            title='Average Lowest Unemployement Rate')

fig.show()
```

```python
fig = px.line(df2, x="Year", y="Unemployement Rate", color='Country Name')
fig.show()
```
